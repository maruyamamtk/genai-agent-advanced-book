## マニュアル検索ツール

```python
import sys, pathlib

# VS Code は __vsc_ipynb_file__ にノートブックの絶対パスをセットする
try:
    root = pathlib.Path(__vsc_ipynb_file__).resolve().parent.parent
except NameError:
    # フォールバック: CWDベースで判定
    p = pathlib.Path().resolve()
    root = p.parent if p.name == "notebooks" else p

if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from src.tools.search_xyz_manual import search_xyz_manual

result = search_xyz_manual.invoke("オンラインヘルプセンター")
print(result[0].content)
```
